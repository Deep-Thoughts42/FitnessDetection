```python
curdir = os.getcwd()

set_path = os.path.join(curdir, "images")
dir_list = os.listdir(set_path)

# Contains every set that we have
dir_list = [item for item in dir_list if item[:3] == "Set"]
```

```python
for dir in dir_list:
    processed_files = os.listdir(os.path.join(set_path, dir, "Processed"))
    for value, image in enumerate(processed_files):
        value += 1
        new_name = dir + "Image" + str(value) + "." + image.split('.')[-1]

        # Now the goal is to resize the images to at least 244 by 244 in order to fit into deeplab
        img = cv2.imread(os.path.join(set_path, dir, "Processed", image)) 
        dim = (224, 224)
        resized = cv2.resize(img, dim)
        new_path = os.path.join(curdir, "images", "ProcessedSets", new_name)
        cv2.imwrite(new_path, resized)

        # Use this to show the images
        # plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
        # plt.show()
        

    
    
```
